```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
tempo = np.arange(0,40,0.01)
serra = tempo%2
serra = [2 - x if x > 1 else x for x in serra]
plt.plot(tempo,serra.)
```

```python
#sinal = tempo/40.0
sinal = 0.5*np.sin(tempo*np.pi/20)+0.5
pwm = (sinal>serra)
plt.plot(tempo,pwm)
plt.fill(tempo,pwm,'c')
plt.plot(tempo,sinal)
```

```python
sinal = np.sin(tempo*np.pi/20)
pwm = [(sin>ser) if sin> 0 else -1*(sin>-ser) for sin,ser in zip(sinal,serra)]
plt.plot(tempo,pwm)
#plt.fill(tempo,pwm,'c')
plt.plot(tempo,sinal)
```
