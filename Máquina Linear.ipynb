# MÃ¡quina Linear

$$F_{ind} = i l B$$
$$F_t = F_{ind}+F_{apl} = m a$$
$$e_{ind} = v B l$$
$$V_B = e_{ind} + i R$$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
VB = 10.0
Fapl = 1.0
R = 0.5
B = 1.0
l = 0.1
v0 = 0
dt = 0.1
T = 500
m = 1
```

```python
index = 0
t0 = 0
t = np.arange(t0,T+t0+dt,dt)
eind = []
i = []
Find = []
v = [v0]
Pel = []
Pmec = []
while index*dt <= T:
    eind.append(v[index]*B*l)
    i.append((VB-eind[index])/R)
    Find.append(i[index]*l*B)
    Ft = Find[index]+Fapl
    a = Ft/m
    Pel.append(eind[index]*i[index])
    Pmec.append(Find[index]*v[index])
    v.append(v[index] + a * dt)
    index = index + 1
v.pop()
```

```python
plt.plot(t,eind)
```

```python
plt.plot(t,i)
```

```python
plt.plot(t,v)
```

```python
plt.plot(t,Find)
```

```python
plt.plot(t,Pmec)
plt.plot(t,Pel)
```
